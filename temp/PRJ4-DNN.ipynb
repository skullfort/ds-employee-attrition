```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf
import pandas as pd 
```

```python
data_df = pd.read_csv("cleaned_mode.csv")
data_df.head()
```

```python
# Determine the number of unique values in each column.
unique_cols = data_df.nunique()
unique_cols
```

```python
final_data_df = pd.get_dummies(data_df)
final_data_df
```

```python
# Split our preprocessed data into our features and target arrays
X = final_data_df.drop(['target'], axis=1)
y = final_data_df['target']

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
number_input_features = len(X_train_scaled[0])
hidden_nodes_layer1 = 80
hidden_nodes_layer2 = 30

nn_model_1 = tf.keras.models.Sequential()

# First hidden layer
nn_model_1.add(tf.keras.layers.Dense(
    units=hidden_nodes_layer1,
    input_dim=number_input_features, 
    activation="relu")
)

# Second hidden layer
nn_model_1.add(tf.keras.layers.Dense(
    units=hidden_nodes_layer2, 
    activation="relu")
)

# Output layer
nn_model_1.add(tf.keras.layers.Dense(
    units=1, 
    activation="sigmoid")
)

# Check the structure of the model
nn_model_1.summary()
```

```python
# Compile the model
nn_model_1.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Train the model
fit_model = nn_model_1.fit(X_train_scaled, y_train, epochs=40)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn_model_1.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
