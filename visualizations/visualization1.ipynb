```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#CSV File
df = pd.read_csv("cleaned_mode.csv")
```

Based on Education

```python
# Extract unique column values and sort them by education level
column_values = df['education_level'].unique()
sorted_values = ['Primary School', 'High School', 'Graduate', 'Masters', 'Phd']
column_mapping = {column_value: column_value.lower().replace(' ', '') for column_value in sorted_values}

# Calculate the total counts and percentages in the sorted order
total_counts = {}
total_with_target_1 = {}
percentages = {}

for column_value, label in column_mapping.items():
    total_counts[label] = df[df['education_level'] == column_value].shape[0]
    total_with_target_1[label] = df[(df['education_level'] == column_value) & (df['target'] == 1)].shape[0]
    # Check if the total count is zero to avoid division by zero
    if total_counts[label] != 0:
        percentages[label] = (total_with_target_1[label] / total_counts[label]) * 100
    else:
        percentages[label] = 0

# Create a bar chart with sorted values
labels = list(column_mapping.values())
values = [percentages[label] for label in labels]

plt.bar(np.arange(len(labels)), values)

# Add percentage values to the bar chart
for i, value in enumerate(values):
    plt.text(i, value, f"{value:.2f}%", ha='center', va='bottom')

# Set the x-axis ticks and labels
plt.xticks(np.arange(len(labels)), labels)

# Set the y-axis label
plt.ylabel('Positive Percentage')

# Set the title of the chart
plt.title('Likelihood of a person leaving based on Education Level')

plt.tight_layout()

# Save the figure
plt.savefig('Basedon_Education.png', bbox_inches='tight')

# Display the chart
plt.show()
```

Based on Company Size

```python
# Manually define the sorting order for company sizes
sorting_order = ['<10', '10-49', '50-99', '100-500', '500-999', '1000-4999', '5000-9999', '>10000']

# Extract unique column values
column_values = df['company_size'].unique()
column_mapping = {column_value: column_value for column_value in column_values}

# Calculate the total counts and percentages in the specified order
total_counts = {}
total_with_target_1 = {}
percentages = {}

for column_value, label in column_mapping.items():
    total_counts[label] = df[df['company_size'] == column_value].shape[0]
    total_with_target_1[label] = df[(df['company_size'] == column_value) & (df['target'] == 1)].shape[0]
    # Check if the total count is zero to avoid division by zero
    if total_counts[label] != 0:
        percentages[label] = (total_with_target_1[label] / total_counts[label]) * 100
    else:
        percentages[label] = 0

# Sort the labels and values based on the specified order
labels_sorted = [label for label in sorting_order if label in column_mapping.keys()]
values_sorted = [percentages[label] for label in labels_sorted]

plt.bar(labels_sorted, values_sorted)

# Add percentage values to the bar chart
for i, value in enumerate(values_sorted):
    plt.text(i, value, f"{value:.2f}%", ha='center', va='bottom')

# Y axis
plt.ylabel('Positive Percentage')

# Title
plt.title('Likelihood of a person leaving based on Company Size')


plt.xticks(rotation=45)
plt.tight_layout()

# Save
plt.savefig('Basedon_CompanySize.png', bbox_inches='tight')

# Display
plt.show()
```

City_Index

```python
# Need to define bin ranges first
bin_ranges = [
    (0, 0.50),
    (0.51, 0.60),
    (0.61, 0.70),
    (0.71, 0.80),
    (0.81, 0.90),
    (0.91, float('inf'))
]

# Create the column mapping dictionary with the bins
column_mapping = {}
for i, (start, end) in enumerate(bin_ranges):
    if end == float('inf'):
        label = f"Greater than {start}"
    else:
        label = f"{start} - {end}"
    column_mapping[i] = label

# Calculate the total counts
total_counts = {}
total_with_target_1 = {}
percentages = {}

for i, (start, end) in enumerate(bin_ranges):
    if end == float('inf'):
        total_counts[column_mapping[i]] = df[df['city_development_index'] > start].shape[0]
        total_with_target_1[column_mapping[i]] = df[(df['city_development_index'] > start) & (df['target'] == 1)].shape[0]
    else:
        total_counts[column_mapping[i]] = df[(df['city_development_index'] >= start) & (df['city_development_index'] <= end)].shape[0]
        total_with_target_1[column_mapping[i]] = df[(df['city_development_index'] >= start) & (df['city_development_index'] <= end) & (df['target'] == 1)].shape[0]

    # Calculate percentages only if the count is not zero
    if total_counts[column_mapping[i]] != 0:
        percentages[column_mapping[i]] = (total_with_target_1[column_mapping[i]] / total_counts[column_mapping[i]]) * 100
    else:
        percentages[column_mapping[i]] = 0

# Create a bar chart
labels = list(column_mapping.values())
values = list(percentages.values())

plt.bar(labels, values)

# Add percentage values to the bar chart
for i in range(len(labels)):
    plt.text(i, values[i], f"{values[i]:.2f}%", ha='center', va='bottom')

# Y axis 
plt.ylabel('Positive Percentage')
plt.xticks(rotation=45)

# Title
plt.title('Likelihood of a person leaving based on city_development_index')

plt.tight_layout()

#Save
plt.savefig('Basedon_CDI1.png', bbox_inches='tight')

# Display
plt.show()
```
